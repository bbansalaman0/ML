```python
from sklearn.model_selection import train_test_split
```

```python
x = train.drop(['id', 'target'], axis=1)
y = train.target
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import MinMaxScaler
import warnings 
warnings.filterwarnings('ignore')
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
regressor = ['LinearRegression', 'DecisionTreeRegressor', 
             'RandomForestRegressor', 'KNeighborsRegressor', 'XGBRegressor']

models = [LinearRegression(), DecisionTreeRegressor(), 
         RandomForestRegressor(), KNeighborsRegressor()]
kf = KFold(n_splits=5, shuffle=True)
```

```python
def model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    on_train_set = model.predict(X_train)
    train_score = np.sqrt(mean_squared_error(y_train, on_train_set))
    on_test_set = model.predict(X_test)
    test_score = np.sqrt(mean_squared_error(y_test, on_test_set))
    print(model)
    print('training_score = ',  train_score)
    print('test_score=',  test_score)
    print('\n')
```

```python
model(models[0], X_train, X_test, y_train, y_test)
```

```python
model(DecisionTreeRegressor(max_depth=5 ), X_train, X_test, y_train, y_test)
```

```python
model(KNeighborsRegressor(n_jobs=-1), X_train, X_test, y_train, y_test)
```

```python
xgb = XGBRegressor(n_jobs = -1,n_estimators=500, max_depth=5 , learning_rate=0.01 )
xgb.fit(X_train, y_train)
```

```python
on_train_set = xgb.predict(X_train)
train_score = np.sqrt(mean_squared_error(y_train, on_train_set))
on_test_set = xgb.predict(X_test)
test_score = np.sqrt(mean_squared_error(y_test, on_test_set))
print(model)
print('training_score = ',  train_score)
print('test_score=',  test_score)
print('\n')
```

```python
model(XGBRegressor(n_jobs = -1,n_estimators=1000, max_depth=8 , learning_rate=0.01 ), X_train, X_test, y_train, y_test)
```

```python
xgb = XGBRegressor(n_jobs = -1,n_estimators=1100, max_depth=8 , learning_rate=0.01 )
xgb.fit(X_train, y_train, early_stopping_rounds=6, eval_set=[(X_test, y_test)], verbose=1)
on_train_set = xgb.predict(X_train)
train_score = np.sqrt(mean_squared_error(y_train, on_train_set))
on_test_set = xgb.predict(X_test)
test_score = np.sqrt(mean_squared_error(y_test, on_test_set))
print(model)
print('training_score = ',  train_score)
print('test_score=',  test_score)
print('\n')
```
