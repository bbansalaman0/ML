### Missing values

#### For categorical NaN values

```python
## Capturing all the NaN values
## 1. Handling for categorical features
categories_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes == 'O']
## printing percentage of nan values in categorical features
for feature in categories_with_nan:
    print('{} : {}% missing values'.format(feature, np.round(dataset[feature].isnull().mean(),4)))
```

##### replacing NaN

```python
## replacing NaN values with new lable

def replace_cat_feature(dataset, feature_nan):
    data = dataset.copy()
    data[feature_nan] = data[feature_nan].fillna('Missing')
    return data
dataset = replace_cat_feature(dataset, categories_with_nan)
dataset[categories_with_nan].isnull().sum()
```

```python
dataset.head()
```

#### For numerical NaN values

```python
## Capturing all the NaN values
## 1. Handling for numerical features

numerical_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes != 'O']

for feature in numerical_with_nan:
    print('{} : {}% missing value'.format(feature, np.round(dataset[feature].isnull().mean(), 4)))
```

##### replacing NaN

```python
## replacing numerical missing values
for feature in numerical_with_nan:
    ## we will replace by using median since there are outliers
    median_value = dataset[feature].median()
    
    ## create a new feature to capture nan values
    dataset[feature + 'nan'] = np.where(dataset[feature].isnull(), 1, 0)
    dataset[feature].fillna(median_value, inplace=True)

dataset[numerical_with_nan].isnull().sum()
```

### numerical values to normal distribution|

#### Conversion 
converting numerical values to normal distributon using log normal diatribution 
converting the numerical feature values those does'nt have 0 value in them

```python
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    dataset[feature] = np.log(dataset[feature])
```

### Handling Rare Categorical variables

we will remove the categorical variables that are present less than 1% of the observation

```python
categories_features = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
```

```python
categories_features
```

```python
for feature in categories_features:
    temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df = temp[temp > 0.1].index
    dataset[feature] = np.where(dataset[feature].isin(temp_df), dataset[feature], 'Rare_var')
```

### Feature Scaling

#### Perform Lable Encoding

```python
from sklearn.preprocessing import LabelEncoder
for feature in categories_features:
    lc = LabelEncoder()
    dataset[feature] = lc.fit_transform(dataset[feature])
```

```python
dataset[categories_features]
```

```python
feature_scale = [feature for feature in dataset.columns if feature not in['Id', 'SalePrice'] ]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
dataset[feature_scale]
```

```python
## transforming the train and test set, and add on the Id and SalePrice variable

data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True), 
                   pd.DataFrame(scaler.transform(dataset[feature_scale]), 
                               columns=feature_scale)], axis= 1)
```
