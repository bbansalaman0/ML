## Checking null features

```python
#checking null values
null_features = [feature for feature in data_train.columns if data_train[feature].isnull().sum()>=1]

for feature in null_features:
    print(feature , np.round((data_train[feature].isnull().mean()), 4), '% missing values')
    
```

```python
#relationship with target variable(SalePrice)

for feature in null_features:
    data = data_train.copy()
    
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

## Temporal features

```python
# relationship of YrSold vs SalePrice

data_train.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Yrsold')
plt.ylabel('SalePrice')
plt.title('YrSold vs SalePrice');
```

```python
# checking differencce b/w YrSold and ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']

for feature in temp_features:
    
    if feature != 'YrSold':
        data = data_train.copy()
        data[feature] = data['YrSold'] - data[feature]
        
        plt.scatter(data[feature], data['SalePrice'],c='red' )
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature + ' vs SalePrice')
        plt.show()
```

## Training dataset with scaling

```python
scaling = MinMaxScaler()

x = train_dataset.drop('SalePrice', axis=1)
x_scaled = scaling.fit_transform(x)

y = train_dataset['SalePrice']
y_scaled = scaling.fit_transform(y.values.reshape(-1,1))
```

## Test dataset

```python
test_dataset.head()
```

```python
test_dataset.drop('SalePrice', axis=1, inplace=True)
```

```python
test_dataset_scaled = scaling.fit_transform(test_dataset)
```

## models

```python
models = {'linreg' : LinearRegression(), 'knn': KNeighborsRegressor(), 
          'dt': DecisionTreeRegressor(), 'rf': RandomForestRegressor(), 'xgb':XGBRegressor(), 'svm': SVR()} 
```

```python
def model(model, xtrain, ytrain, ytest):
    cv = cross_val_score(model, xtrain, ytrain, cv=5, n_jobs = -1)
    model.fit(xtrain, ytrain)
    y_pred = model.predict(ytest)
    score = mean_squared_error(ytrain, y_pred)
    print(model, ': scores: ',cv, '\nmeanscores:', cv.mean(),'root_mean_squared_error: ',np.sqrt(score), '\nmean_squared_error: ',score  )
    
    print('\n')
    
    
    
    
    
```

```python
for i, j in models.items():
    model(j, x_scaled,y_scaled , test_dataset_scaled )
```

from above scores RandomForestRegressor() and XGBRegressor() have performed well
